```python
#3d plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
theta = np.linspace(-3*np.pi,3*np.pi,200)
z = np.linspace(-3,3,200)
r =z**3 +1
x = r*np.sin(theta)
y = r*np.cos(theta)
ax.plot(x,y,z,label='Parametric curve')
ax.legend()
plt.show()
```

# Steps to create your own ufuncs functions

- Use a def Keyword to define a funtion
- Add created function to Numpy ufunc library using the frompyfuns()
- Call this function over a NumPy array

```python
integers = np.arange(1,101)
integers
```

```python
def modulo(val):
    return (val % 10)

#adding into numpy
mod_10 = np.frompyfunc(modulo,1,1)
mod_int = mod_10(integers)

print("Mod Integers:",mod_int)
```
