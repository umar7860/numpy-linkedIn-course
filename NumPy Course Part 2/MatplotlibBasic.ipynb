```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
average_monthly_temperature = [39.1,40.1,48.0,50.4,60.3,73.7,80.0,76.9,68.8,57.9,53.0,39.2]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
fig = plt.figure()
plt.plot(months,average_monthly_temperature)
#for title
plt.title("Average Monthly Temperature")
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.show
```

```python
#Saving graph
fig.savefig('graph.png')
fig.savefig('graph.pdf')
```

# Understanding figures

```python
x = np.arange(3)
plt.plot(x,x)
plt.plot(x,2*x)
plt.plot(x,3*x)
plt.grid(True)
plt.show
```

```python
x = np.linspace(0,5,5)
y = 2*x
plt.plot(x,y)
#plt.show
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
plt.show()
```

# Matplotlib subplots functionality

```python
fig = plt.figure()
x = np.arange(3)
y = 2 * x
#2 rows , 2 colums ,  1 plot
plt.subplot(2,2,1)
plt.plot(x,y,'b')

plt.subplot(2,2,2)
plt.plot(x,1-y,'r')

plt.subplot(2,2,3)
plt.plot(x,2-y,'g')

plt.subplot(2,2,4)
plt.plot(x,y,'y')

plt.show()
```

```python
fig , axs = plt.subplots(2,2, figsize = (6,6))
axs[0,0].plot(x,y,'b')
axs[0,1].plot(x,1-y,'r')
axs[1,0].plot(x,2-y,'g')
axs[1,1].plot(x,y,'y')
```

# Understanding_legends

```python
#Legend is the description of each graph
x = np.linspace(1,10)
first_line = plt.plot(x,x+1,label = 'y=x+1')
plt.legend()
```

```python
second_line = plt.plot(x,x+2,linestyle='solid',label = 'y=x+2')
#second_line.set_label('y=x+2')
third_line = plt.plot(x,x+3,linestyle='dashed',label = 'y=x+3')
#third_line.set_label('y=x+3')
plt.legend()
```

```python
first_plot = plt.plot([1,2,3],label='first_plot')
second_plot = plt.plot([3,2,1],label='second_plot')
third_plot = plt.plot([2,2,2],label='third_plot')
plt.legend(bbox_to_anchor=(1.02,1.0),borderaxespad = 0);
```
