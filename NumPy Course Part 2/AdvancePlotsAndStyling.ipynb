# Advanced Matplotlib Commands

```python
x = np.linspace (1,10,1024)
plt.xscale('log')
plt.yscale('log')
plt.plot(x,x,label = '$f(x)=x$')
plt.plot(x,10**x,label = '$f(x)=10^x$')
plt.plot(x,np.log(x),label = '$f(x)=log(x)$')
plt.legend()
plt.show()
```

```python
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
x = np.arange(0.0,50.0,0.1)
y = x**2

fig , ax = plt.subplots()
ax.plot(x,y)
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.xaxis.set_minor_formatter('{x:.0f}')

ax.xaxis.set_minor_locator(MultipleLocator(2))
plt.show()
```

```python
x = np.arange(0.0,50.0,0.1)
y = x**2

fig , ax = plt.subplots()
ax.plot(x,y)
ax.set_xlim([0,50])
ax.set_ylim([0,2500])
plt.show()
```

```python
x = np.arange(0.0,50.0,0.1)
y = x**2

fig , ax = plt.subplots()
ax.plot(x,y)

list_1 = list(range(5,55,5))
ax.set_xticks(list_1)
list_2 = list(range(0,2500,250))
ax.set_yticks(list_2)
plt.show()
```

# Adding annotations

```python
x = np.linspace(0,10)
y1 = x
y2 = 8-x
fig , ax = plt.subplots()
plt.plot(x,y1,label='Supply')
plt.plot(x,y2,label='Demand')
ax.annotate("Equilibrium",xy=(4,4),xytext=(3,2),\
            fontsize=12,fontweight='semibold',\
            arrowprops =dict(linewidth=2,arrowstyle='->'))
#label for x-axes
plt.xlabel('quantity',fontsize=12)
#label for y-axes
plt.ylabel('price',fontsize=12)

plt.legend()
plt.show()
```

```python
x = np.linspace(0,10)
y1 = x
y2 = 8-x
fig , ax = plt.subplots()
plt.plot(x,y1,label='Supply')
plt.plot(x,y2,label='Demand')

bbox_props = dict(boxstyle='rarrow',fc = (0.8,0.9,0.9),lw=2)
t = ax.text(2,4,"equilibrium", ha="center",va="center",rotation=0,
           size=10,bbox=bbox_props)
#label for x-axes
plt.xlabel('quantity',fontsize=12)
#label for y-axes
plt.ylabel('price',fontsize=12)

plt.legend()
plt.show()
```

# Figures

```python
from matplotlib.patches import Circle,Polygon
from matplotlib.collections import PatchCollection
fig , ax = plt.subplots()
patches= [ ]

#draw circle and triangle

circle = Circle((.42,.75),0.12)
triangle = Polygon([[0.1,0.5],[0.2,0.7],[0.3,0.54]],True)
patches += [circle,triangle]
#Draw the patches

colors = 100*np.random.rand(len(patches)) 
p = PatchCollection(patches)
p.set_array(np.array(colors))
ax.add_collection(p)
#show the figure
plt.show()
```

# Pie and Bar charts in Numpy

```python
#Pie Chart
preffered_workoption = [10.7,47.6,38.8,2.9]
colors = ['b','g','r','c']
explode=(0,0.2,0,0)
labels = ['Collected','Hybrid','Fully Remote','Not applicable']


plt.pie(preffered_workoption,colors = colors,labels = labels,
       explode=explode,autopct='%1.1f%%',
        counterclock=False,shadow=True)
plt.title('Preffered workoption')
plt.show()
```

```python
#Bar chart
preffered_workoption = [10.7,47.6,38.8,2.9]
colors = ['b','g','r','c']
#explode=(0,0.2,0,0)
widths=[0.7,0.7,0.7,0.7]
labels = ['Collected','Hybrid','Fully Remote','Not applicable']
plt.bar(range(0,4),preffered_workoption,width=widths,color=colors,align='center')
plt.title('Preffered workoption')
plt.show()
```
